## Question 2

```python
np.random.seed(42)

n = 17
p = 0.5
size = 70
x3 = np.random.binomial(n=n,p=p,size=size)

#count how many times we got 1, 2,...17 heads
head_count = [np.equal(x3, i).sum() for i in range(n)]  # list comprehension
print(head_count)

probs70 = [np.equal(x3, i).mean() for i in range(n)] # .mean calcs probability
print(probs70)
```

```python
plt.plot(list(range(n)), probs70, color='blue', marker='o')
plt.xlabel("Number of heads")
plt.ylabel("Probability")
```
